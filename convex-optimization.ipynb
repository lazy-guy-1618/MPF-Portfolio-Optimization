```python
import cvxpy as cp
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
daily_ret_matrix = df1.pct_change(-1).dropna()
# expected daily return and covariance matrix
daily_r_vec = daily_ret_matrix.mean(axis=0)
daily_r_cov = daily_ret_matrix.cov()
# turn into annual data
r_annualized = daily_r_vec * 252
cov_annualized = np.array(daily_r_cov * 252)
```

```python
(daily_ret_matrix + 1).iloc[::-1].cumprod().iloc[::-1]
```

```python
ret = []
std = []

num_simulation = 10000
num_funds = len(daily_r_vec)

W = np.random.normal(1, 3, (num_simulation, num_funds))
W[W<0] = 0 # long only

for i in range(num_simulation):
    w = W[i]
    if sum(w):
    # set the weight vector elementarily sum to 1
        w = w * (1 / np.sum(w))
    else:
        w = np.random.uniform(0, 1, (num_funds,))
        w = w * (1 / np.sum(w))
    
    ret.append(w @ r_annualized)
    std.append((w @ cov_annualized @ w.T) ** 0.5)
    
ret = np.array(ret)
std = np.array(std)

# the current 10 years Treasury rate as the risk free rate (rf)
rf = 0.0339
sharpe = (ret - rf) / std

print('Desriptive summary of the sharpe ratios of the randomly generated portfolios')
print(pd.Series(sharpe).describe())

print('Best performance among the randomized weights\n')
print(f'  Sharpe Ratio: {np.max(sharpe):.4f}')
print(f'  Annualized Expected Return (%): {ret[np.argmax(sharpe)]:.4f}')
print(f'  Annualized Risk (%): {std[np.argmax(sharpe)]:.4f}')

plt.figure()
plt.scatter(std, ret)
plt.ylabel('Expected Return (%)', fontsize=14)
plt.xlabel('Risk (%)', fontsize=14)
plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```

```python
# our expected min return is 0.09% per year
r_min_annualized = 0.09

# create cvxpy variable to minimize
w = cp.Variable(num_funds)
r_min = r_min_annualized / 252

# construct the objective function and constraints
obj = cp.Minimize(w.T @ cov_annualized @ w)
const = [
    cp.sum(w) == 1, w >= 0,
    w.T @ r_annualized- r_min >= 0
]

# solve it!
prob = cp.Problem(obj, const)
opt_v = prob.solve()

risk_opt = (opt_v * 252) ** 0.5
w_opt = w.value

print('optimal risk (%):', risk_opt)
print('optimal proportion:', np.round(w_opt, 4))
```

# Pareto Optimality

```python
num_portfolios = 10000
results = np.zeros((3, num_portfolios))
for i in range(num_portfolios):
    weights = np.random.random(15)
    weights /= np.sum(weights)
    portfolio_return = np.sum(daily_r_vec * weights)
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(daily_r_cov, weights)))
    results[:, i] = [portfolio_return, portfolio_std_dev, portfolio_return / portfolio_std_dev]
```

```python
from scipy.optimize import minimize
num_assets = 15
# Function to minimize
def minimize_volatility(weights):
    return np.sqrt(np.dot(weights.T, np.dot(daily_r_cov, weights)))

# Constraints and bounds
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})  # The weights must sum to 1.
bounds = tuple((0, 1) for asset in range(num_assets))

# Array to store results
frontier_y = np.linspace(0.01, 0.03, 200)  # Example range of returns
frontier_x = []

for possible_return in frontier_y:
    cons = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1},
            {'type': 'eq', 'fun': lambda x: np.sum(x * daily_r_vec) - possible_return})
    result = minimize(minimize_volatility, num_assets*[1./num_assets,], method='SLSQP', bounds=bounds, constraints=cons)
    frontier_x.append(result['fun'])

# Plot the Efficient Frontier
plt.plot(frontier_x, frontier_y, 'r--', linewidth=3)
```

```python
import matplotlib.pyplot as plt

plt.scatter(results[1, :], results[0, :], c=results[2, :], cmap='YlGnBu', marker='o')
plt.colorbar(label='Sharpe Ratio')
plt.plot(frontier_x, frontier_y, 'r--', linewidth=3)  # Efficient Frontier
plt.title('Efficient Frontier')
plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Expected Return')
plt.xlim(0.0,0.4)
plt.ylim(0.00,0.0150)
plt.show()
plt.savefig('pareto.png')
```
